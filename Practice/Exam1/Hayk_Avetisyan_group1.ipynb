## Selecting best combination of features

```python
df3 = pd.read_csv('car_data.csv')
df3 = df3[['wheel_base', 'length', 'width', 'height', 'curb_weight','horsepower', 'price']]
```

```python
df3 = df3.fillna(method='bfill')
```

```python
df3.isna().sum()
```

```python
import itertools as it
from sklearn.model_selection import train_test_split
from sklearn import linear_model 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

pd.options.display.max_rows=1000

regression_num = 0
metrics = pd.DataFrame(columns=['Num_Regr', 'RMSE_trn', 'RMSE_tst', 'RMSE_diff', 'R2_trn', 'R2_tst', 'cf_det_R2'])

reg_data = pd.DataFrame(columns=['Num_Regr', 'b0']+list(df3.columns.difference(['price'])))


for k in range(1,len(df3.columns.difference(['price']))+1):
    regressor_groups = it.combinations(df3.columns.difference(['price']),k)
    for regressors in regressor_groups:
        #print("Regressors=",regressors,'\n')
        
        xTrain, xTest, yTrain, yTest = train_test_split(df3[list(regressors)],df3[['price']], test_size=0.2, random_state=42)
       
        regr = linear_model.LinearRegression()
        regr.fit(xTrain, yTrain)
        coef_determ_R2 = regr.score(xTrain,yTrain)
        b0, b = regr.intercept_, regr.coef_
        
        #Train regr
        y_pred_trn = regr.predict(xTrain)
        rmse_trn = np.sqrt(mean_squared_error(y_pred_trn, yTrain))
        r2_trn = r2_score(y_pred_trn, yTrain)
        
        #Test regr
        y_pred_tst = regr.predict(xTest)
        rmse_tst = np.sqrt(mean_squared_error(y_pred_tst, yTest))
        r2_tst = r2_score(y_pred_tst, yTest)
        
        metrics.loc[regression_num,'Num_Regr'] = regression_num
        metrics.loc[regression_num,'RMSE_trn'] = rmse_trn
        metrics.loc[regression_num,'RMSE_tst'] = rmse_tst
        metrics.loc[regression_num,'R2_trn'] = r2_trn
        metrics.loc[regression_num,'R2_tst'] = r2_tst
        metrics.loc[regression_num,'cf_det_R2'] = coef_determ_R2
       
        reg_data.loc[regression_num, 'Num_Regr'] = regression_num
        reg_data.loc[regression_num,'b0'] = b0[0]
        reg_data.loc[regression_num, list(regressors)] = b
        reg_data.replace({np.NaN:0}, inplace=True)
             
        regression_num += 1
        
metrics['RMSE_diff'] = metrics['RMSE_tst'].subtract(metrics['RMSE_trn'])

print('\n')
print('metrics for k regressions:\n')
print(metrics)
print('\n')
print("Regressors included in k-th regression:")
reg_data
```
