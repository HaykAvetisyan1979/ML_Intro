## Load 'Iris.csv' dataset, perform exploratory analysis(draw box plots, bar plots of features, scatter plots of feature pairs)

```python
df = pd.read_csv('Iris.csv')
df.head()
```

## Box plots

```python
df.boxplot(column=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
features = df[df.columns.difference(['Id','Species'])]
fig, ax = plt.subplots(len(features.columns),1)
fig.set_figwidth(15)
fig.set_figheight(25)
fig.tight_layout(h_pad=5)

for i, feature in enumerate(features.columns.tolist()):
    ax[i].hist(df[feature])
    ax[i].set_title(feature)
```

```python
df[df.columns.difference(['Id','Species'])].corr()
```

```python
import itertools as it
pairs = it.combinations(features.columns,2)

fig, ax = plt.subplots(6,1)
fig.set_figwidth(10)
fig.set_figheight(45)
fig.tight_layout(h_pad=5)

for i, pair in enumerate(pairs):
    ax[i].scatter(df[pair[0]],df[pair[1]])
    ax[i].set_title(str(pair[0])+" - "+ str(pair[1])) 
    ax[i].set_xlabel(str(pair[0]))
    ax[i].set_ylabel(str(pair[1]))                 
```

## 1. Logistic regression

In the dataset there are three species of iris: 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'. Process the dataset and split it with 80/20 ratio. Fit two logistic regression models:
1. The first model must predict whether an entry is Iris-setosa or not.
2. The second model must predict whether an entry is Iris-virginica or not.

Using this two models, predict classes of test set and report accuracy.

Make 2 scatter plots using all data, both train and test.
1. Plot SepalLengthCm vs SepalWidthCm, color according to their classes, plot the decision boundaries of the two models on the same plot.
2. Plot PetalLengthCm vs PetalWidthCm, color according to their classes, plot the decision boundaries of the two models on the same plot.


```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.Species.replace({'Iris-setosa':1,'Iris-versicolor':0,'Iris-virginica':0},inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest1, yTrain, yTest1 = train_test_split(df[df.columns.difference(['Id','Species'])], df[['Species']], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(C=0.01, solver='liblinear')
logr.fit(xTrain, yTrain)
y_pred_lgr1=logr.predict(xTest1)
```

```python
y_pred_lgr1
```

```python
from sklearn import metrics
cnf_matrix1 = metrics.confusion_matrix(yTest1, y_pred_lgr1)
cnf_matrix1
```

```python
print("Accuracy:",metrics.accuracy_score(yTest1, y_pred_lgr1))
print("Precision:",metrics.precision_score(yTest1, y_pred_lgr1))
print("Recall:",metrics.recall_score(yTest1, y_pred_lgr1))
```

## 2. Kmeans

Use the splitted dataset in task 1 to fit a Kmeans model on training set with K = 3. Infer which cluster center corresponds to which class. Predict classes on test set and report accuracy, compare it with previous task. Repeat the two plots from the previous task. Add the cluster centers to the plot.

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.Species.replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3},inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(df[df.columns.difference(['Id','Species'])], df[['Species']], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(xTrain)
xTransf_trn = ss.transform(xTrain)
xTransf_tst = ss.transform(xTest)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=100)
kmeans.fit(xTransf_trn)
y_kmeans = kmeans.predict(xTransf_tst)
y_kmeans
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yTest, y_kmeans)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(yTest, y_kmeans)
```
